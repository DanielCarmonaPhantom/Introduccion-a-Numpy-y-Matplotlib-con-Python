## **Ejercicio 9**
Crea una gráfica de la ecuación $f(x)=(x+1)^{3}$, con los datos de X ya dados y que cumplan los siguientes parámetros:

- cambia su color a verde "green" 
- colocale el título "Mi gráfica"
- nombra al eje horizontal "Eje X" 
- nombra al eje vertical "Eje Y"
- cambia su marcador a: D
- coloca el límite de visibilidad de la grafica de (-15 a 15) en el eje X y de (-70 a 70) en el eje Y
- coloca una malla o cuadrícula en la gráfica de color "blue" y estilo de línea: --

Crea otra gráfica de la ecuación $f(x)=(x+1)^{2}$ en la misma cuadrícula y debe cumplir con los siguientes parámetros:

- cambia su color a magenta "magenta" 
- cambia su marcador a: o
- cambia su estilo de línea a: -.

Para diferenciar las gráficas, coloca un Nombre de serie, la ecuación $f(x)=(x+1)^{3}$ llevará por nombre "Ecuación cúbica" y la ecuación $f(x)=(x+1)^{2}$, llevará por nombre "Ecuación cuadrática"

Finalmente coloca un sombreado de color amarillo claro (#FDFF7C) en la intersección, es decir, en las coordenadas del rectángulo dado por los puntos: p1(-5,25), p2(5,25), p3(5,-25), p4(-5,-25)

El resultado del ejercicio debera verse exactamente igual a lo siguiente:

Ej3.JPG

```python
x=np.arange(-50,50,0.5)
# YOUR CODE HERE
y1 = (x+1)**3
y2 = (x+1)**2

plt.xlim(-15,15)
plt.ylim(-100,100)

plt.grid(color='blue', linestyle='--')

plt.title("Mi gráfica")

plt.xlabel("Eje X")
plt.ylabel("Eje Y")



plt.plot(x,y1, color = 'green', marker = 'D', label= 'Ecuación cúbica')
plt.plot(x,y2, color = 'magenta', marker = 'o', linestyle='-.', label= 'Ecuación cuadrática')
plt.legend()

px = [-5, 5, 5, -5]
py = [25, 25, -25, -25]

plt.fill(px,py , color='cyan') 

# NO ELIMINAR LA SIGUIENTE LINEA DE CODIGO
plt.savefig('ejercicio9_respuesta.png', dpi = 100)
plt.show()
```

## **Ejercicio 10**

Separa las gráficas del ejercicio anterior con todas las características como el título, nombre de los ejes. cuadrícula, color de las gráficas, marcador, nombre de serie. Lo único que debes de quitar es el sombreado.

Añade en la gráfica de  $f(x)=(x+1)^{3}$, el texto: " f(x)=(x+1)^{3}" en la coordenada (-10,55) y colocalo de color rojo

Añade en la gráfica de  $f(x)=(x+1)^{2}$, el texto: " f(x)=(x+1)^{2}" en la coordenada (-5,-40) y colocalo de color rojo

El resultado del ejercicio debera verse exactamente igual a lo siguiente:

image.png
