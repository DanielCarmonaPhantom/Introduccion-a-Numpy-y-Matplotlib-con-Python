Crea otra grfica de la ecuacin $f(x)=(x+1)^{2}$ en la misma cuadrcula y debe cumplir con los siguientes parmetros:

- cambia su color a magenta "magenta" 
- cambia su marcador a: o
- cambia su estilo de lnea a: -.

Para diferenciar las grficas, coloca un Nombre de serie, la ecuacin $f(x)=(x+1)^{3}$ llevar por nombre "Ecuacin cbica" y la ecuacin $f(x)=(x+1)^{2}$, llevar por nombre "Ecuacin cuadrtica"

Finalmente coloca un sombreado de color amarillo claro (#FDFF7C) en la interseccin, es decir, en las coordenadas del rectngulo dado por los puntos: p1(-5,25), p2(5,25), p3(5,-25), p4(-5,-25)

## **Ejercicio 10**

Separa las grficas del ejercicio anterior con todas las caractersticas como el ttulo, nombre de los ejes. cuadrcula, color de las grficas, marcador, nombre de serie. Lo nico que debes de quitar es el sombreado.

Aade en la grfica de  $f(x)=(x+1)^{3}$, el texto: " f(x)=(x+1)^{3}" en la coordenada (-10,55) y colocalo de color rojo

Aade en la grfica de  $f(x)=(x+1)^{2}$, el texto: " f(x)=(x+1)^{2}" en la coordenada (-5,-40) y colocalo de color rojo
